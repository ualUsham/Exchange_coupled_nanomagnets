```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('data.xlsx')
data.head()
```

```python
correlation_matrix = data.corr()

plt.figure(figsize=(8, 6))

# Create heatmap with color scaling and annotations
plt.imshow(correlation_matrix, cmap='coolwarm')  # Adjust colormap as desired
plt.colorbar()  # Add a colorbar to interpret color intensity

# Set labels for rows and columns (optional)
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Add annotations (optional) to show correlation values directly on the heatmap
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        plt.text(i, j, f"{correlation_matrix.iloc[i, j]:.2f}", ha='center', va='center', fontsize=8)

# Adjust layout (optional)
plt.tight_layout()

# Display the heatmap
plt.show()
```

```python
input_features = list(data.iloc[:, :-1])



# Loop through each feature and create a scatter plot
for feature in input_features:
    plt.figure(figsize=(4,4))
    plt.scatter(data[feature], data['Hn'])
    plt.xlabel(feature)
    plt.ylabel('Hn')
    plt.title(f"{feature} vs. Output Feature")
    plt.tight_layout()  # Adjust spacing to avoid overlapping elements
    plt.show()
```
